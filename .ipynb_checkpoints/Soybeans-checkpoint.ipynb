## Frequency Distributions of the Categorical Features

```python
# Let’s plot the distribution of each feature
def plot_distribution(dataset, cols=5, width=20, height=15, hspace=0.2, wspace=0.5):
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure(figsize=(width,height))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=wspace, hspace=hspace)
    rows = math.ceil(float(dataset.shape[1]) / cols)
    for i, column in enumerate(dataset.columns):
        ax = fig.add_subplot(rows, cols, i + 1)
        ax.set_title(column)
        if dataset.dtypes[column] == np.object:
            g = sns.countplot(y=column, data=dataset)
            substrings = [s.get_text()[:18] for s in g.get_yticklabels()]
            g.set(yticklabels=substrings)
            plt.xticks(rotation=25)
        else:
            g = sns.distplot(dataset[column])
            plt.xticks(rotation=25)
            
cols_to_plot = ['class', 'date', 'plant_stand', 'precip', 'temp', 'hail', 'crop_hist', \
           'area_damaged', 'severity', 'seed_tmt', 'germination', 'plant_growth', \
          'leaves', 'leafspots_halo', 'leafspots_marg', 'leafspot_size', 'leaf_shread', \
           'leaf_malf', 'leaf_mild', 'stem', 'lodging', 'stem_cankers', 'canker_lesion',\
          'fruiting_bodies', 'external_decay', 'mycelium', 'int_discolor', 'sclerotia',\
          'fruit_pods', 'fruit_spots', 'seed', 'mold_growth', 'seed_discolor', 'seed_size',\
          'shriveling', 'roots']
plot_distribution(df[cols_to_plot], cols=3, width=20, height=20, hspace=0.45, wspace=0.5)
```

## Handle Missing Data

Below we chose to use the median to replace the missing data. 

```python
df.date = pd.to_numeric(df.date, errors='ignore')
df.date = df.date.fillna(int(pd.to_numeric(df.date,errors="coerce").median(skipna = True)))
df.plant_stand = pd.to_numeric(df.plant_stand, errors='ignore')
df.plant_stand = df.plant_stand.fillna(int(pd.to_numeric(df.plant_stand,errors="coerce").median(skipna = True)))
df.precip = pd.to_numeric(df.precip, errors='ignore')
df.precip = df.precip.fillna(int(pd.to_numeric(df.precip,errors="coerce").median(skipna = True)))
df.temp = pd.to_numeric(df.temp, errors='ignore')
df.temp = df.temp.fillna(int(pd.to_numeric(df.temp,errors="coerce").median(skipna = True)))
df.hail = pd.to_numeric(df.hail, errors='ignore')
df.hail = df.hail.fillna(int(pd.to_numeric(df.hail,errors="coerce").median(skipna = True)))
df.crop_hist = pd.to_numeric(df.crop_hist, errors='ignore')
df.crop_hist = df.crop_hist.fillna(int(pd.to_numeric(df.crop_hist,errors="coerce").median(skipna = True)))
df.area_damaged = pd.to_numeric(df.area_damaged, errors='ignore')
df.area_damaged = df.area_damaged.fillna(int(pd.to_numeric(df.area_damaged,errors="coerce").median(skipna = True)))
df.severity = pd.to_numeric(df.severity, errors='ignore')
df.severity = df.severity.fillna(int(pd.to_numeric(df.severity,errors="coerce").median(skipna = True)))
df.seed_tmt = pd.to_numeric(df.seed_tmt, errors='ignore')
df.seed_tmt = df.seed_tmt.fillna(int(pd.to_numeric(df.seed_tmt,errors="coerce").median(skipna = True)))
df.germination = pd.to_numeric(df.germination, errors='ignore')
df.germination = df.germination.fillna(int(pd.to_numeric(df.germination,errors="coerce").median(skipna = True)))
df.plant_growth = pd.to_numeric(df.plant_growth, errors='ignore')
df.plant_growth = df.plant_growth.fillna(int(pd.to_numeric(df.plant_growth,errors="coerce").median(skipna = True)))
df.leaves = pd.to_numeric(df.leaves, errors='ignore')
df.leaves = df.leaves.fillna(int(pd.to_numeric(df.leaves,errors="coerce").median(skipna = True)))
df.leafspots_halo = pd.to_numeric(df.leafspots_halo, errors='ignore')
df.leafspots_halo = df.leafspots_halo.fillna(int(pd.to_numeric(df.leafspots_halo,errors="coerce").median(skipna = True)))
df.leafspots_marg = pd.to_numeric(df.leafspots_marg, errors='ignore')
df.leafspots_marg = df.leafspots_marg.fillna(int(pd.to_numeric(df.leafspots_marg,errors="coerce").median(skipna = True)))
df.leafspot_size = pd.to_numeric(df.leafspot_size, errors='ignore')
df.leafspot_size = df.leafspot_size.fillna(int(pd.to_numeric(df.leafspot_size,errors="coerce").median(skipna = True)))
df.leaf_shread = pd.to_numeric(df.leaf_shread, errors='ignore')
df.leaf_shread = df.leaf_shread.fillna(int(pd.to_numeric(df.leaf_shread,errors="coerce").median(skipna = True)))
df.leaf_malf = pd.to_numeric(df.leaf_malf, errors='ignore')
df.leaf_malf = df.leaf_malf.fillna(int(pd.to_numeric(df.leaf_malf,errors="coerce").median(skipna = True)))
df.leaf_mild = pd.to_numeric(df.leaf_mild, errors='ignore')
df.leaf_mild = df.leaf_mild.fillna(int(pd.to_numeric(df.leaf_mild,errors="coerce").median(skipna = True)))
df.stem = pd.to_numeric(df.stem, errors='ignore')
df.stem = df.stem.fillna(int(pd.to_numeric(df.stem,errors="coerce").median(skipna = True)))
df.lodging = pd.to_numeric(df.lodging, errors='ignore')
df.lodging = df.lodging.fillna(int(pd.to_numeric(df.lodging,errors="coerce").median(skipna = True)))
df.stem_cankers = pd.to_numeric(df.stem_cankers, errors='ignore')
df.stem_cankers = df.stem_cankers.fillna(int(pd.to_numeric(df.stem_cankers,errors="coerce").median(skipna = True)))
df.canker_lesion = pd.to_numeric(df.canker_lesion, errors='ignore')
df.canker_lesion = df.canker_lesion.fillna(int(pd.to_numeric(df.canker_lesion,errors="coerce").median(skipna = True)))
df.fruiting_bodies = pd.to_numeric(df.fruiting_bodies, errors='ignore')
df.fruiting_bodies = df.fruiting_bodies.fillna(int(pd.to_numeric(df.fruiting_bodies,errors="coerce").median(skipna = True)))
df.external_decay = pd.to_numeric(df.external_decay, errors='ignore')
df.external_decay = df.external_decay.fillna(int(pd.to_numeric(df.external_decay,errors="coerce").median(skipna = True)))
df.mycelium = pd.to_numeric(df.mycelium, errors='ignore')
df.mycelium = df.mycelium.fillna(int(pd.to_numeric(df.mycelium,errors="coerce").median(skipna = True)))
df.int_discolor = pd.to_numeric(df.int_discolor, errors='ignore')
df.int_discolor = df.int_discolor.fillna(int(pd.to_numeric(df.int_discolor,errors="coerce").median(skipna = True)))
df.sclerotia = pd.to_numeric(df.sclerotia, errors='ignore')
df.sclerotia = df.sclerotia.fillna(int(pd.to_numeric(df.sclerotia,errors="coerce").median(skipna = True)))
df.fruit_pods = pd.to_numeric(df.fruit_pods, errors='ignore')
df.fruit_pods = df.fruit_pods.fillna(int(pd.to_numeric(df.fruit_pods,errors="coerce").median(skipna = True)))
df.fruit_spots = pd.to_numeric(df.fruit_spots, errors='ignore')
df.fruit_spots = df.fruit_spots.fillna(int(pd.to_numeric(df.fruit_spots,errors="coerce").median(skipna = True)))
df.seed = pd.to_numeric(df.seed, errors='ignore')
df.seed = df.seed.fillna(int(pd.to_numeric(df.seed,errors="coerce").median(skipna = True)))
df.mold_growth = pd.to_numeric(df.mold_growth, errors='ignore')
df.mold_growth = df.mold_growth.fillna(int(pd.to_numeric(df.mold_growth,errors="coerce").median(skipna = True)))
df.seed_discolor = pd.to_numeric(df.seed_discolor, errors='ignore')
df.seed_discolor = df.seed_discolor.fillna(int(pd.to_numeric(df.seed_discolor,errors="coerce").median(skipna = True)))
df.seed_size = pd.to_numeric(df.seed_size, errors='ignore')
df.seed_size = df.seed_size.fillna(int(pd.to_numeric(df.seed_size,errors="coerce").median(skipna = True)))
df.shriveling = pd.to_numeric(df.shriveling, errors='ignore')
df.shriveling = df.shriveling.fillna(int(pd.to_numeric(df.shriveling,errors="coerce").median(skipna = True)))
df.roots = pd.to_numeric(df.roots, errors='ignore')
df.roots = df.roots.fillna(int(pd.to_numeric(df.roots,errors="coerce").median(skipna = True)))
```
